```python
import matplotlib.pyplot as plt
import numpy as np

# Define the range for n (must be positive)
n = np.linspace(0.1, 10, 1000)

# Define different values of d to show multiple curves
d_values = [1, 2, 5, 10, 50]

# Create the plot
plt.figure(figsize=(10, 6))

for d in d_values:
    # Calculate y = n^(-1/d) = 1/n^(1/d)
    y = n ** (-1 / d)
    plt.plot(n, y, label=f"d = {d}", linewidth=2)

plt.xlabel("n", fontsize=12)
plt.ylabel("n^(-1/d)", fontsize=12)
plt.title("Plot of f(n) = n^(-1/d) = 1/ᵈ√n for different values of d", fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.xlim(0, 10)
plt.ylim(0, 3)

# Add some annotations
plt.text(
    8,
    2.5,
    "As d increases,\nthe curve becomes\nflatter",
    bbox=dict(boxstyle="round,pad=0.3", facecolor="lightblue", alpha=0.7),
)

plt.tight_layout()
plt.show()

# Optional: Create a 3D surface plot showing how the function varies with both n and d
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection="3d")

# Create meshgrid for 3D plot
n_3d = np.linspace(0.1, 10, 50)
d_3d = np.linspace(1, 5, 50)
N, D = np.meshgrid(n_3d, d_3d)

# Calculate Z = N^(-1/D)
Z = N ** (-1 / D)

# Create surface plot
surf = ax.plot_surface(N, D, Z, cmap="viridis", alpha=0.8)

ax.set_xlabel("n")
ax.set_ylabel("d")
ax.set_zlabel("n^(-1/d)")
ax.set_title("3D Surface: f(n,d) = n^(-1/d)")

# Add colorbar
fig.colorbar(surf)

plt.tight_layout()
plt.show()
```
